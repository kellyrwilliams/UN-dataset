```python
gdp_subset = gdp_df.drop(columns = 'Value Footnotes')
gdp_rename = gdp_subset.rename(columns = {'Country or Area': 'Country', 'Value': 'GDP_Per_Capita'})
gdp_rename
```

```python
internet_subset = internet_df.drop(columns = 'Value Footnotes')
internet_rename = internet_subset.rename(columns = {'Country or Area':'Country', 'Value':'Internet_Users_Pct'})
internet_rename
```

```python
gdp_and_internet_use = pd.merge(gdp_rename, internet_rename,
                        left_on=['Country', 'Year'],
                        right_on=['Country', 'Year'],
                          how='outer',
                        indicator=True)
gdp_and_internet_use.head(10)
```

```python
gdp_internet_subset = gdp_and_internet_use.query('Year in [2004, 2009, 2014]')
gdp_internet_subset
```

```python
gdp_internet_2004 = gdp_internet_subset.query('Year == 2004')
gdp_internet_2004.head(5)
```

```python
gdp_internet_2009 = gdp_internet_subset.query('Year == 2009')
gdp_internet_2009.head(5)
```

```python
gdp_internet_2014 = gdp_internet_subset.query('Year == 2014')
gdp_internet_2014.head(5)
```

```python
gdp_internet_2014.sort_values('Internet_Users_Pct', ascending=False).head(5)
```
